# Accessing and Changing specific elements, rows, columns

```python
a = np.array([[1,2,3,4,5,6,7],[5,6,7,8,9,10,11]])
print(a)
```

```python
# Accessing 9 [r,c]
print(a[1,4])
print(a[-1,-3])
```

```python
# Get a specific Row
a[0,:]
```

```python
# Specific Column
a[:,3]
```

```python
# Getting a sequence of data 
a[:,0:7:2]
```

```python
a[1,4] = 40
print(a)

a[:,4] = 70
print(a)

a[:,4] = [10,20]
print(a)
```

```python
b = np.array([[[1,2],[2,9]],[[3,4],[4,5]]])
print(b)
```

```python
# Getting some specific element
b[0,1,1]
```

# Different Types of Arrays

```python
# All zeros matrix
np.zeros(5)
```

```python
a = np.zeros((2,3))
```

```python
# All 1's matrix
np.ones((2,3),dtype = 'int16')
```

```python
np.full((2,2),100)
```

```python
np.full(a.shape,100)
```

```python
# Random Decimal Numbers
np.random.random_sample(a.shape)
```

```python
# Random Integer Numbers
np.random.randint(4,7,size = (3,3))
```

```python
np.identity(5)
```

```python
arr = np.copy(a)
print(arr)
arr_3 = np.repeat(a, 3, axis=0)
print(arr_3)
```

```python
a = np.ones((5,5))
b = np.zeros((3,3))
b[1,1] = 9
print(b.shape)
a[1:4,1:4] = b
print(b)
print(a)
```

```python
a = np.array([1,2,3])
b = a 
print(b)
b[0] = 12
print(a)
print(b)
```

```python
b = np.copy(a)
b[0] = 50
print(a)
print(b)
```
