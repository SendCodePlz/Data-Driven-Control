```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris(as_frame=True)
X_iris = iris.data[["petal length (cm)", "petal width (cm)"]].values
Y_iris = iris.target

tree_clf = DecisionTreeClassifier(max_depth=2,random_state=42)
tree_clf.fit(X_iris,Y_iris)
```

```python
from sklearn.tree import export_graphviz

dot_data = export_graphviz(
        tree_clf,
        out_file="iris_tree.dot",
        feature_names=["petal length (cm)","petal width (cm)"],
        class_names=iris.target_names,
        rounded=True,
        filled=True
)

```

```python
from graphviz import Source

Source.from_file("iris_tree.dot")
```

```python
from sklearn.datasets import make_moons

X_moons, y_moons = make_moons(n_samples=150, noise=0.2, random_state=42)

tree_clf1 = DecisionTreeClassifier(random_state=42)
tree_clf2 = DecisionTreeClassifier(min_samples_leaf=5, random_state=42)
tree_clf1.fit(X_moons, y_moons)
tree_clf2.fit(X_moons, y_moons)
```

```python
import matplotlib.pyplot as plt

X_moons, y_moons = make_moons(n_samples=150, noise=0.2, random_state=42)

plt.figure(figsize=(6,6))
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=y_moons, edgecolor="k")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Two Moons Dataset")
plt.show()
```

```python
X_moons_test, y_moons_test = make_moons(n_samples=1000, noise=0.2,random_state=43)
print(tree_clf1.score(X_moons_test,y_moons_test))
print(tree_clf2.score(X_moons_test,y_moons_test))
```

```python
# Regression

from sklearn.tree import DecisionTreeRegressor

rng = np.random.default_rng(seed=42) # Better Random number generator
X_quad = rng.random((200,1)) - 0.5
y_quad = X_quad ** 2 + 0.025 * rng.standard_normal((200, 1))

tree_reg = DecisionTreeRegressor(random_state=42,max_depth=2)
tree_reg.fit(X_quad,y_quad)
```

```python
# Two trees
tree_depth2 = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_depth5 = DecisionTreeRegressor(max_depth=5, random_state=42)

tree_depth2.fit(X_quad, y_quad)
tree_depth5.fit(X_quad, y_quad)

# Dense input for smooth plotting
X_plot = np.linspace(-0.5, 0.5, 1000).reshape(-1, 1)

y_pred_2 = tree_depth2.predict(X_plot)
y_pred_5 = tree_depth5.predict(X_plot)

# Plot
plt.figure(figsize=(10,6))

plt.scatter(X_quad, y_quad, s=15, alpha=0.5, label="Data")
plt.plot(X_plot, y_pred_2, label="max_depth=2", linewidth=3)
plt.plot(X_plot, y_pred_5, label="max_depth=5", linewidth=2)

plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.title("Decision Tree Regression")
plt.show()
```

```python
# Sensitivity to Axis Orientation

from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

pca_pipeline = make_pipeline(StandardScaler(),PCA())
X_iris_rotated = pca_pipeline.fit_transform(X_iris)
tree_clf_pca = DecisionTreeClassifier(max_depth=2, random_state=42)
tree_clf_pca.fit(X_iris_rotated, Y_iris) 
```

```python
# ---------- Plotting function ----------
def plot_decision_boundary(clf, X, y, ax):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 300),
        np.linspace(y_min, y_max, 300)
    )

    X_new = np.c_[xx.ravel(), yy.ravel()]
    y_pred = clf.predict(X_new)
    y_pred = y_pred.reshape(xx.shape)

    ax.contourf(xx, yy, y_pred, alpha=0.3)
    ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor="k", s=30)
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)

# ---------- Plot ----------
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

plot_decision_boundary(tree_clf, X_iris, Y_iris, axes[0])
axes[0].set_title("Original Feature Space (Depth = 2)")

plot_decision_boundary(tree_clf_pca, X_iris_rotated, Y_iris, axes[1])
axes[1].set_title("After PCA Rotation (Depth = 2)")

plt.show()
```
